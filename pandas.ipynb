```python
# cells with data of worng farmat .
#comvert into a correct format 
import pandas as pd

df = pd.read_csv('data.csv')

df['Calories'] = pd.to_datetime(df['Calories'], format='mixed')

print(df.to_string())
```

```python
# delete the nah / null cells 
import pandas as pd

df = pd.read_json('data.json')
df.dropna(subset=['Calories'], inplace = True)

print(df.to_string())
```
